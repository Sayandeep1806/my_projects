```python
df = pd.read_html(url)
df
```

```python
for item in df:
    BS = item[0].astype(str).str.contains('Retained') | item[0].astype(str).str.contains('Total Assets') | item[0].astype(str).str.contains('Current Assets')
    if BS.any():
        Balance_Sheet = item

Balance_Sheet
```

```python
Balance_Sheet = Balance_Sheet.iloc[1:,[1,4,10]]
Balance_Sheet
```

```python
#grab the first row for the header
header = Balance_Sheet.iloc[0]
Balance_Sheet = Balance_Sheet[2:] #take the data less the header row
Balance_Sheet.columns = header
Balance_Sheet.columns.values[0] = 'Item'
Balance_Sheet
```

```python
def symbol_handling(DF,cols):
    for col in cols:
        DF[DF.columns[col:]] = DF[DF.columns[col:]].astype(str)

        DF[DF.columns[col]] = DF[DF.columns[col]].map(lambda x: x.replace('(','-'))
        DF[DF.columns[col]] = DF[DF.columns[col]].map(lambda x: x.replace(')',''))
        DF[DF.columns[col]] = DF[DF.columns[col]].map(lambda x: x.replace(',',''))
        DF[DF.columns[col]] = DF[DF.columns[col]].map(lambda x: x.replace('—','nan'))
        
    DF[DF.columns[1:]] = DF[DF.columns[1:]].astype(float)
    return DF
```

```python
symbol_handling(Balance_Sheet,[1,2])
```
