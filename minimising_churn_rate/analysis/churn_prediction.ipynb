Minimising customer churn (i.e., users who cancel their subscription) of subscription product through analysis of financial habits/ customer behavior. We primarily aim to re-engage the users with the product by incorporating new features to the product, (which might of interest to such customers) and making the potential churn customers aware of such features during our marketing campains like email, etc.

```python
# reading the dataset
dataset = pd.read_csv(r'../data/new_churn_data.csv')
```

## Data Preparation

```python

user_identifier = dataset['user']
dataset = dataset.drop(columns = ['user']) # unused for prediction purposes
```

```python
# One-Hot Encoding - converting categorical variables to independent binary columns, as categorical variables cannot be interpreted as such
dataset.housing.value_counts()
```

```python
dataset = pd.get_dummies(dataset) 
# get_dummies does one-hot encoding for us. automatically detects categorical variables and converts them to columns for us
```

```python
dataset.columns
```

```python
# avoiding the dummy variable trap, which happens when we consider all the categorical variables associated with a column
# trick to avoid dummy variable trap: consider one less one-hot encoded column for each categorical feature as the result would be obvious

dataset = dataset.drop(columns = ['housing_na', 'zodiac_sign_na', 'payment_type_na'])
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset.drop(columns = 'churn'), dataset['churn'],
                                                    test_size = 0.2,
                                                    random_state = 0)
```

```python
y_train.value_counts()
```

```python
# Balancing the Training Set - for scenarios where the model is biased but still gives a high accuracy rate (> 50%)
# in this case we try to balance the training set,i.e., balance amount of 0's and 1's that training set has in its response variable
# in this way we can guarantee that whatever lift or accuracy we have within our model is a result of the model itself and not a random permutation/combination of the 0s and 1s present in the training set

pos_index = y_train[y_train.values == 1].index
neg_index = y_train[y_train.values == 0].index

if len(pos_index) > len(neg_index):
    higher = pos_index
    lower = neg_index
else:
    higher = neg_index
    lower = pos_index

random.seed(0)
higher = np.random.choice(higher, size=len(lower))
lower = np.asarray(lower)
new_indexes = np.concatenate((lower, higher))

X_train = X_train.loc[new_indexes,]
y_train = y_train[new_indexes]
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler() # generates a np array which needs to be converted to a df to prevent the loss of index
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
X_test2 = pd.DataFrame(sc_X.transform(X_test))
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
X_train2.index = X_train.index.values
X_test2.index = X_test.index.values
X_train = X_train2
X_test = X_test2
```

## Model Building

```python
# Fitting Model to the Training Set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicting Test Set
y_pred = classifier.predict(X_test)
```

```python
# Evaluating Results
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
print('Accuracy:',accuracy_score(y_test, y_pred))
print('Precision:',precision_score(y_test, y_pred)) # tp / (tp + fp)
print('Recall:',recall_score(y_test, y_pred)) # tp / (tp + fn)
print('f1-score:',f1_score(y_test, y_pred))
```

```python
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)
sn.heatmap(df_cm, annot=True, fmt='g',cmap='Blues')
print("Test Data Accuracy: %0.4f" % accuracy_score(y_test, y_pred))
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Logistic Regression Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2)) # mean-accuracy value
```

```python
# Analyzing Coefficients
pd.concat([pd.DataFrame(X_train.columns, columns = ["features"]),
           pd.DataFrame(np.transpose(classifier.coef_), columns = ["coef"])
           ],axis = 1)
```

## Feature Selection

```python
# Recursive Feature Elimination - effective at selecting those features (columns) in a training dataset that are more or most relevant in predicting the target variable.
# two important configuration options when using RFE: the choice in the number of features to select and the choice of the algorithm used to help choose features. 
from sklearn.feature_selection import RFE
```

```python
# Model to Test
classifier = LogisticRegression()

# Select Best X Features
rfe = RFE(classifier, 20)
rfe = rfe.fit(X_train, y_train)

# summarize the selection of the attributes
print(rfe.support_) # tells about the columns considered in the final result
```

```python
# Display the columns with rankings based on importance
table=[]
for i in range(len(X_train.columns[rfe.support_])):
    table.append([X_train.columns[rfe.support_][i],rfe.ranking_[i]])
    
display(pd.DataFrame(table,columns=['Column','Rank']).sort_values(by=['Rank']).reset_index(drop=True))
```

```python
# New Correlation Matrix
sn.set(style="white")

# Compute the correlation matrix
corr = X_train[X_train.columns[rfe.support_]].corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(18, 15))

# Generate a custom diverging colormap
cmap = sn.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sn.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}) 
```

```python
# Fitting Model to the new Training Set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train[X_train.columns[rfe.support_]], y_train)

# Predicting new Test Set
y_pred = classifier.predict(X_test[X_train.columns[rfe.support_]])
```

```python
# Evaluating Results
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)

print('Accuracy:',accuracy_score(y_test, y_pred))
print('Precision:',precision_score(y_test, y_pred)) # tp / (tp + fp)
print('Recall:',recall_score(y_test, y_pred)) # tp / (tp + fn)
print('f1-score:',f1_score(y_test, y_pred))
```

```python
df_cm = pd.DataFrame(cm, index = (1, 0), columns = (1, 0))
plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)
sn.heatmap(df_cm, annot=True, fmt='g',cmap='Blues')
print("Test Data Accuracy: %0.4f" % accuracy_score(y_test, y_pred))
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier,
                             X = X_train[X_train.columns[rfe.support_]],
                             y = y_train, cv = 10)
print("Logistic Regression Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2))
```

```python
# Analyzing Coefficients
pd.concat([pd.DataFrame(X_train[X_train.columns[rfe.support_]].columns, columns = ["features"]),
           pd.DataFrame(np.transpose(classifier.coef_), columns = ["coef"])
           ],axis = 1)
```
